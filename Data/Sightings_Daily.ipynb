```python
#Imports
import os
import pandas as pd
import numpy as np
from datetime import datetime, date
```

```python
#Read CSV
ufo_data = pd.read_csv('../UFO_data.csv')
```

```python
#Remove American Territories - US States Only
ufo_data = ufo_data.drop(ufo_data[ufo_data['State'] == "American Samoa"].index)
```

```python
#Switch to Datetime
ufo_data['Date'] = pd.to_datetime(ufo_data['Date'])
ufo_data
```

```python
ufo_data.dtypes
```

```python
#Get Hour of Day (UTC)
ufo_data["Date"].dt.hour
```

```python
#Add Columns
ufo_data["Hour"] = ufo_data["Date"].dt.hour
ufo_data["Month"] = ufo_data["Date"].dt.month_name()
ufo_data["Day"] = ufo_data["Date"].dt.day_name()
```

```python
#Drop Columns
ufo_daily = ufo_data.drop(['Date','Month', 'State','Day','City', 'Country', 'Shape', 'Duration', 'Summary', 'Posted','Images', 'Location', 'Latitude', 'Longitude'], axis=1, inplace=True)
ufo_data
```

```python
ufo_sightings = ufo_data[['Hour', 'ReportNum']]
```

```python
#Group Total Sightings by Time of Day
print("Count of UFO Sightings by Time of Day:")
ufo_daily = ufo_sightings.groupby('Hour').count()['ReportNum'].to_frame(name = 'Sightings').reset_index()
ufo_daily
```

```python
# write to csv file
ufo_daily.to_csv('sightings_time.csv')
```
